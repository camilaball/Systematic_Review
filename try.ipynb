```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Methylation Method': ['Pyrosequencing', 'Sequencing', 'Array'],
    'Imputation Used': [2, 14, 26],
    'No Imputation': [20, 18, 92]
}

df = pd.DataFrame(data)

#Methylation Method' is index
df.set_index('Methylation Method', inplace=True)

#plotting
ax = df.plot(kind='bar', stacked=True, figsize=(8, 5), color=['#5DADE2', '#AED6F1'])
plt.ylabel('Number of Clocks')
plt.title('Imputation Use by Methylation Measurement Method')
plt.xticks(rotation=0)
plt.legend(title='Imputation Status')
plt.tight_layout()


plt.savefig('imputation_per_methylation_platorm.pdf')
plt.show()
```
